# POLYNOMIAL REGRESSION

## IMPORT THE LIBRARIES

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## IMPORTING DATASETS

```
datasets = pd.read_csv('Position_Salaries.csv')
x = datasets.iloc[:,1:-1].values
y = datasets.iloc[:,-1].values
```

```
print(x)
```

```
print(y)
```

## TRAINING THE LINEAR REGRESSION MODEL ON THE WHOLE DATASETS

```
from sklearn.linear_model import LinearRegression
Lin_reg1 = LinearRegression()
Lin_reg1.fit(x, y)
```

## TRAINING THE POLYNOMIAL REGRESSION MODEL ON THE WHOLE DATASETS

```
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
x_poly = poly_reg.fit_transform(x)
Lin_reg2 = LinearRegression()
Lin_reg2.fit(x_poly, y)
```

## VISUALLISION OF LINEAR REGRESSION RESULT

```
plt.scatter(x, y, color = 'red')
plt.plot(x, Lin_reg1.predict(x), color = 'blue')
plt.title('LINEAR REGRESSION RESULT')
plt.xlabel('LEVEL OF THE PERSON')
plt.ylabel('SALARY OF THE PERSON')
plt.show()
```

## VISUALLISION OF POLYNOMIAL REGRESSION RESULT

```
plt.scatter(x, y, color = 'red')
plt.plot(x, Lin_reg2.predict(x_poly), color = 'blue')
plt.title('POLYNOMIAL REGRESSION RESULT')
plt.xlabel('LEVEL OF THE PERSON')
plt.ylabel('SALARY OF THE PERSON')
plt.show()
```

## VISUALLISING THE POLYNOMIAL REGRESSION RESULT (FOR HIGHER RESOLUTION AND SMOOTHER CURVE)

```
x_grid = np.arange(min(x), max(x), 0.1)
x_grid = x_grid.reshape(len(x_grid), 1)
```

```
plt.scatter(x, y, color = 'red')
plt.plot(x_grid, Lin_reg2.predict(poly_reg.fit_transform(x_grid)), color = 'blue')
plt.title('POLYNOMIAL REGRESSION RESULT WITH SMOOTH')
plt.xlabel('LEVEL OF THE PERSON')
plt.ylabel('SALARY OF THE PERSON')
plt.show()
```

## PREDICTING A NEW RESULT WITH LINEAR REGRESSION 

```
Lin_reg1.predict([[6.5]])
```

## PREDICTING A NEW RESULT WITH POLYNOMIAL REGRESSION

```
Lin_reg2.predict(poly_reg.fit_transform([[6.5]]))
```
